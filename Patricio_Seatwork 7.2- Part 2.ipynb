```python
import pandas as pd

# Load the meteorite data
meteorites = pd.read_csv("Meteorite_Landings.csv")

# Display the first few rows
meteorites.head()
```

```python
# Convert 'year' to datetime and extract the numerical year
meteorites["year"] = pd.to_datetime(meteorites["year"], errors="coerce").dt.year

# Display the updated dataframe
meteorites.head()
```

```python
# Filter data for years 2005-2009
filtered_meteorites = meteorites[(meteorites["year"] >= 2005) & (meteorites["year"] <= 2009)]

# Display the first few rows of the filtered data
filtered_meteorites.head()
```

```python
# Create a pivot table with count and 95th percentile of mass grouped by year and fall type
pivot_table = filtered_meteorites.pivot_table(
    index="year",
    columns="fall",
    values="mass (g)",
    aggfunc=["count", lambda x: x.quantile(0.95)]
)

# Rename columns for clarity
pivot_table.columns = ["Count_Fell", "Count_Found", "95th_Percentile_Fell", "95th_Percentile_Found"]

# Display the pivot table
print(pivot_table)
```

```python
# Generate summary statistics for mass based on fall type
summary_stats = meteorites.groupby("fall")["mass (g)"].describe()

# Display summary statistics
print(summary_stats)
```

```python
# Load the taxi trip data
taxi_data = pd.read_csv("2019_Yellow_Taxi_Trip_Data.csv", parse_dates=["tpep_dropoff_datetime"])

# Display the first few rows
taxi_data.head()
```

```python
# Resample to an hourly frequency based on dropoff time
hourly_summary = taxi_data.resample("H", on="tpep_dropoff_datetime").agg({
    "trip_distance": "sum",
    "fare_amount": "sum",
    "tolls_amount": "sum",
    "tip_amount": "sum"
})

# Display the first few rows
print(hourly_summary.head())
```

```python
# Find the 5 hours with the most tips
top_5_hours = hourly_summary.nlargest(5, "tip_amount")

# Display the top 5 hours
print(top_5_hours)
```
